```python
# Valores dos animais
boi = 250.00
vaca = 216.00
ovelha = 202.50
cabra = 202.50
porco = 101.00

# Dados adicionais
porcentagem_racao = {
    "boi": 0.03,      # 3% do peso vivo por dia
    "vaca": 0.03,     # 3% do peso vivo por dia
    "ovelha": 0.16,   # 16% do peso vivo por dia
    "cabra": 0.05,    # 5% do peso vivo por dia
    "porco": 3        # 3 kg por dia
}

while True:
    nome_proprietario = input('Digite seu nome: ')

    tipo_animal = input('Digite o tipo do animal entre: (boi, vaca, ovelha, cabra, porco) ').strip().lower()
    peso = float(input('Digite o peso do animal em kg: '))

    # Solicitar o valor da ração por kg
    custo_racao_por_kg = float(input('Digite o valor da ração por kg: '))

    if tipo_animal in porcentagem_racao:
        # Calcular o valor do animal
        if tipo_animal == "boi":
            valor = peso / 15 * boi
            tipo_animal_nome = "boi"
        elif tipo_animal == "vaca":
            valor = peso / 15 * vaca
            tipo_animal_nome = "vaca"
        elif tipo_animal == "ovelha":
            valor = peso / 15 * ovelha
            tipo_animal_nome = "ovelha"
        elif tipo_animal == "cabra":
            valor = peso / 15 * cabra
            tipo_animal_nome = "cabra"
        elif tipo_animal == "porco":
            valor = peso / 15 * porco
            tipo_animal_nome = "porco"

        # Calcular a quantidade de ração diária e o custo diário
        if tipo_animal == "porco":
            racao_diaria = porcentagem_racao[tipo_animal]  # em kg
        else:
            racao_diaria = peso * porcentagem_racao[tipo_animal]  # kg de ração por dia

        custo_diario = racao_diaria * custo_racao_por_kg

        # Se o animal for uma vaca, calcular o lucro
        if tipo_animal == "vaca":
            litros_leite_por_dia = float(input('Quantos litros de leite a vaca produz por dia? '))
            preco_por_litro = float(input('Qual o preço do litro de leite? '))
            receita_diaria = litros_leite_por_dia * preco_por_litro
            lucro_diario = receita_diaria - custo_diario

            # Imprimir resultados para a vaca
            print(f'O valor do seu {tipo_animal_nome} é {valor:.2f}')
            print(f'A quantidade de ração diária para o seu {tipo_animal_nome} é {racao_diaria:.2f} kg')
            print(f'O custo diário de ração para o seu {tipo_animal_nome} é {custo_diario:.2f}')
            print(f'A receita diária com leite é {receita_diaria:.2f}')
            print(f'O lucro diário da vaca é {lucro_diario:.2f}')
        else:
            # Imprimir resultados para outros animais
            print(f'O valor do seu {tipo_animal_nome} é {valor:.2f}')
            print(f'A quantidade de ração diária para o seu {tipo_animal_nome} é {racao_diaria:.2f} kg')
            print(f'O custo diário de ração para o seu {tipo_animal_nome} é {custo_diario:.2f}')

        print('Assim que possível, atualizaremos o sistema com melhorias de informação.')

    else:
        print('Escolha um animal entre os selecionados na tela.')

    # Pergunta se o usuário deseja continuar
    continuar = input('Deseja calcular o valor de outro animal? (s/n): ').strip().lower()
    if continuar != 's':
        break

























```
